```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# Load dataset
DATA_PATH = os.path.join("..", "data", "day.csv")
df = pd.read_csv(DATA_PATH)
```

```python
# Summary
print(df.info())
print(df.describe())
```

```python
# Plot rental counts over time
plt.figure(figsize=(12, 5))
plt.plot(df["dteday"], df["cnt"], label="Total Rentals")
plt.title("Daily Bike Rentals Over Time")
plt.xlabel("Date")
plt.ylabel("Count")
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# Seasonality by weekday
df["weekday"] = df["weekday"].astype("category")
sns.boxplot(x="weekday", y="cnt", data=df)
plt.title("Rental Distribution by Weekday")
plt.show()
```
